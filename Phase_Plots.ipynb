```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.colors as colors
#from matplotlib import cm
import matplotlib.ticker as ticker
import sys
import data_importerDM as di
```

```python
post_liquid = np.array([[0.7, 2.5], [0.3, 2.5], \
                             [0.2, 5], [0.2, 15], \
                            [0.3, 15], [0.4, 25], \
                            [0.5, 15], [0.6, 15], \
                            [0.7, 10]]) 
post_stripe = np.array([[0.7, 10], [0.6, 15],\
                        [0.5, 15], [0.55,20],\
                       [0.6, 20], [0.6, 30]])
```

```python
# Phase plot

plt.rc('font', size=30)

if __name__ == "__main__":

    index = 1
    rows=2
    columns=2

    liquid_fd = post_liquid[:,1]/10
    liquid_np = post_liquid[:,0]
    stripe_fd = post_stripe[:,1]/10
    stripe_np = post_stripe[:,0]
    
    gs=gridspec.GridSpec(rows,columns)
    fig = plt.figure(figsize=(6*columns,6*rows))

    ax1 = fig.add_subplot(gs[:])  #scatter plot of particles

    ax1.set_xlabel("Drive Force/Post Force")
    ax1.set_ylabel("Post Density")   
    #ax1.set_title("Posts Phase Plot")
    ax1.set_xlim(0,3.0)
    ax1.set_ylim(0,0.7)
    ax1.set_aspect(5)
    
    ax1.plot(liquid_fd, liquid_np, color = "k")
    ax1.plot(stripe_fd, stripe_np, color = "k")
    
    #ax1.legend(loc = "best", fontsize = 12)
```
